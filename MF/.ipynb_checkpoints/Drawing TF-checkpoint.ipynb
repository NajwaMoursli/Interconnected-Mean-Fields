### Draw TF

```python
TF1_test=np.zeros((201,201))
TF2_test=np.zeros((201,201))
w0=2e-10

for ve in np.linspace(0,200,201):
    for vi in np.linspace(0,200,201):
        #for w in np.linspace(0,200,21):
            
            w=vi*params['b'][1]
            
            index_ve=int(ve)
            index_vi=int(vi)
            #index_w=int(w/10)
            
            #print(ve,vi)
            
            TF1_test[index_ve][index_vi]=TF[0](ve,vi,0)
            TF2_test[index_ve][index_vi]=TF[1](ve,vi,w)
            
            #print('\n \n INPUTS = ',ve,vi,'\n','\n')
```

```python
fig2=plt.figure(figsize=(20,12))
ax21=fig2.add_subplot(221)
ax22=fig2.add_subplot(222)



#heatmap = np.array(firingrates_MF_exc).reshape(len(fetrues_list), len(finhibs_list))

im21=ax21.imshow(TF1_test, cmap='hot', interpolation='nearest')
#ax21.set_yticklabels(fetrues_list)
ax21.set_ylabel("Excitatory Frequencies")
#ax.set_xticklabels(finhibs_list)
#ax21.set_xlim(0,100)
ax21.invert_yaxis()
ax21.set_xlabel("Inhibitory Frequencies")
ax21.set_title("Transfer function RS cell")
fig2.colorbar(im21, ax=ax21)

im22=ax22.imshow(TF2_test, cmap='hot', interpolation='nearest')
#ax.set_yticklabels(fetrues_list)
ax22.set_ylabel("Excitatory Frequencies")
#ax.set_xticklabels(finhibs_list)
ax22.invert_yaxis()
ax22.set_xlabel("Inhibitory Frequencies")
ax22.set_title("Transfer function FS cell")
fig2.colorbar(im22,ax=ax22)
```
